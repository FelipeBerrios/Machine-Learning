# Bike Sharing: Predicción de Demanda Horaria

$ \textbf{Descripción del problema:} $
Los sistemas de intercambio de bicicletas son un medio para alquilar bicicletas en las que el proceso de obtención de membresía, alquiler y retorno de bicicletas se automatiza a través de una red de ubicaciones de quioscos en toda la ciudad. Utilizando estos sistemas, la gente puede alquilar una bicicleta desde una ubicación y devolverla a un lugar diferente según sea necesario. Actualmente, hay más de 500 programas para compartir bicicletas en todo el mundo.


Los datos generados por estos sistemas los hacen atractivos para los investigadores debido a que se registran explícitamente la duración del viaje, la ubicación de salida, la ubicación de llegada y el tiempo transcurrido. Por lo tanto, los sistemas de compartimiento de bicicletas funcionan como una red de sensores, que puede utilizarse para estudiar la movilidad en una ciudad. En esta competición, se pide a los participantes que combinen los patrones de uso histórico con los datos meteorológicos para predecir la demanda de alquiler de bicicletas en el programa Capital Bikeshare en Washington, DC.

La función es altamente no lineal y no determinista como función de la hora del día.

# Objetivo

Entrenar un modelo para obtener un puntaje correspondiente al top-100 del "leaderboard”, es decir
superior o igual a 0.37748. La función utilizada para evaluar este concurso Kaggle se proporciona en la
siguiente ecuación:

$$ E_{bikes}(y, \hat{y}) = \frac{1}{n} \sum_{i} (ln(y_{i} + 1) - ln(\hat{y_i} + 1))^{2}$$

Donde $y,\hat{y} \in {\rm I\!R^{n}}$ denotan los vectores de observaciones y predicciones respectivamente.

# B) Árbol de predicción con parámetros por defecto

A continuación se crea la matriz de entrenamiento y pruebas según el número de datos de cada conjunto. Para las columnas que corresponden a los atributos o variables del modelo, se utilizan las siguientes: $\textit{season, holiday, workingday, weather, temp, atemp, humidity, windspeed, hour}$. La variable $\textit{datetime}$ no es utilizada debido a que esta formateada y es de tipo string, lo que no puede ser procesado en el modelo. Las variables $\textit{count}$, $\textit{casual}$ y $\textit{registered}$ no pueden ser utilizadas ya que influyen directamente en la variable objetivo.

Se crea un árbol de regresión con parámetros por defecto, para evaluar el desempeño tanto en el set de entrenamiento y pruebas. Notar que aquí la evaluación sobre los modelos se realiza con la función entregada por el problema, que no es equivalente al método de score que traen los algoritmos, que para arboles de regresión  es el coeficiente de determinación $R^{2}$.

Para evaluar que tan buenas son las predicciones y visualizarlas gráficamente, se crea una figura que compara las predicciones con los valores reales.

```python
from sklearn.tree import DecisionTreeRegressor as Tree
def eval_bikemodel(y_predict,y_true):
    diff = np.log(y_predict+1.0) - np.log(y_true+1.0)
    return np.sqrt(np.sum(np.square(diff))/len(y_predict))

Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour']]
Xdf2=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed']]

Ydf=df.ix[:,'count']
X_train = Xdf[0:ntrain].values
X_val = Xdf[ntrain:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:ntrain].values
Y_val = Ydf[ntrain:nval].values
Y_test = Ydf[nval:].values

model = Tree(random_state=0)
model.fit(X_train,Y_train)
score_test = model.score(X_test,Y_test)
print "SCORE TEST=%f"%score_test

Y_pred_train = model.predict(X_train)
Y_pred_val = model.predict(X_val)
Y_pred_test = model.predict(X_test)
kagg_train = eval_bikemodel(Y_pred_train,Y_train)
kagg_val = eval_bikemodel(Y_pred_val,Y_val)
kagg_test = eval_bikemodel(Y_pred_test,Y_test)
print "KAGG EVAL TRAIN =%f"%kagg_train
print "KAGG EVAL VAL =%f"%kagg_val
print "KAGG EVAL TEST =%f"%kagg_test

X_train2 = Xdf2[0:ntrain].values
X_test2 = Xdf2[nval:].values
model.fit(X_train2,Y_train)
Y_pred_test2 = model.predict(X_test2)

plt.figure(1, figsize=(20, 6))
plt.subplot(121)
plt.xlabel("Y Test")
plt.ylabel("Y Prediction")
plt.title("Comparacion prediccion vs valor real")
plt.plot(Y_test,Y_pred_test,'.')

plt.subplot(122)
plt.xlabel("Y Test")
plt.ylabel("Y Prediction")
plt.title("Comparacion prediccion vs valor real sin Hora")
plt.plot(Y_test, Y_pred_test2, '.')
plt.show()
```

Como primera aproximación, el árbol de regresión entrega buenos resultados considerando el hecho de que no se tiene que configurar ninguna propiedad. Otro rasgo interesante de los arboles es que no requieren pre procesamiento de los datos como normalización o introducir variables "dummy". También la generación del árbol es muy rápida, pero esto es un detalle de la implementación de sklearn, ya que en general crear un árbol no balanceado puede llevar un tiempo considerable dependiendo del número de datos.

El primer problema que se observa, es que el árbol sobre ajusta demasiado, pasando de un KAGG error de entrenamiento de 0.028516, a 0.574539 en el set de pruebas. Esto puede deberse principalmente a que el árbol se expande en profundidad hasta que cada nodo hoja tenga un subconjunto puro, provocando que sea demasiado complejo, generando así reglas demasiado específicas.

Con respecto al gráfico, se debe notar que si la predicción fuera completamente correcta, se formaría una línea correspondiente a $ y=x$. Se aprecia que para valores pequeños, el árbol predice muy bien, ya que se forma una línea muy agrupada de puntos. Esto indica según los histogramas anteriormente revisados, que el árbol predice muy bien para valores donde tiene una gran cantidad de datos, pero no es capaz de expandir este comportamiento para áreas menores de información, como se observa para valores mayores de YTest.

El grafico de la derecha principalmente describe la importancia del atributo $\textit{hour}$, ya que sin él, el árbol genera un modelo muy malo y no es capaz de predecir adecuadamente. Es por esto que el atributo $\textit{hour}$ es tan relevante para el análisis, así como otros atributos que se puedan obtener a partir del atributo $\textit{datetime}$.

# C) Mejora del árbol de regresión

Para mejorar el árbol de regresión, es necesario modificar sus "parámetros" como puede ser su máxima profundidad, mínimo numero de datos en las hojas o también el mínimo numero de datos requeridos para hacer una división de un nodo interno. Esto es muy importante, ya que estos modelos tienden a sobre-ajustar, entonces al limitar la máxima profundidad lo que se busca es generalizar en mayor medida, ya que quizás se pierdan reglas específicas y no se incluyan algunos atributos en el modelo, pero de esta forma el modelo no será tan dependiente de los datos de entrenamiento. Con respecto al mínimo numero de datos en los nodos hojas, si es muy pequeño, puede que el modelo sobre ajuste, por lo mismo que se mencionó anteriormente, la profundidad del árbol. Si por el contrario el mínimo numero de datos en las hojas es muy grande provocara el efecto contrario, es decir, que los nodos no se puedan dividir y por ende no se creen reglas, haciendo que el modelo no aprenda de los datos.

A continuación se muestra la evaluación de estos parámetros en el set de validación, considerando múltiples valores, para luego escoger los que den mejores resultados y así utilizarlos para obtener un modelo que mejore su desempeño sobre los datos de prueba.

```python
kagg_list_val = []
for i in range(2,31):
    model = Tree(random_state=0,max_depth=i)
    model.fit(X_train,Y_train)
    Y_pred_val = model.predict(X_val)
    kagg_val = eval_bikemodel(Y_pred_val,Y_val)
    kagg_list_val.append(kagg_val)

iaux = kagg_list_val.index(min(kagg_list_val))   
best_depth = [iaux, kagg_list_val[iaux]]   

plt.figure(1, figsize=(20, 6))
plt.subplot(121)
plt.xlabel("Max depth")
plt.ylabel("Kagg Error")
plt.title("Kagg Error en funcion de la profundidad del arbol")
plt.plot(range(2,31),kagg_list_val)

kagg_list_val = []
for i in range(1,50):
    model = Tree(random_state=0,min_samples_leaf=i)
    model.fit(X_train,Y_train)
    Y_pred_val = model.predict(X_val)
    kagg_val = eval_bikemodel(Y_pred_val,Y_val)
    kagg_list_val.append(kagg_val)

iaux = kagg_list_val.index(min(kagg_list_val))    
best_leaf = [iaux, kagg_list_val[iaux]] 
    
plt.subplot(122)
plt.xlabel("Min Samples leaf")
plt.ylabel("Kagg Error")
plt.title("Kagg Error en funcion del minimo numero de ejemplos en nodo hoja")
plt.plot(range(1,50),kagg_list_val)
plt.show()

#Kagg test
model = Tree(random_state=0,min_samples_leaf=best_leaf[0]+1, max_depth=best_depth[0]+2)
model.fit(X_train,Y_train)
Y_pred_test = model.predict(X_test)
kagg_val = eval_bikemodel(Y_pred_test,Y_test)


print "BEST KAGG EVAL VAL =%f, DEPTH: %f "%(best_depth[1],best_depth[0]+2)
print "BEST KAGG EVAL VAL =%f, leaf_samples: %f "%(best_leaf[1],best_depth[0]+1)

print "BEST KAGG EVAL TEST =%f "%(kagg_val)
```

En este caso se analiza la profundidad dejando los datos de los nodos hojas por defecto, y luego se varían los datos mínimos en los nodos hojas, dejando la profundidad por defecto.

Los resultados muestran que ambos cambios, mejoran significativamente el KAGG error en el set de validación, pasando de 0.553973 a 0.476274. Luego al unir ambos resultados, se genera un modelo para evaluar el testing set, el cual da un resultado de 0.493951, y que anteriormente era de 0.574539 sin ajuste de parámetros.

Esto muestra la importancia de utilizar ajuste de los parámetros de los árboles de decisión, ya que al ser modelos simples, pueden sobre ajustar y una manera de prevenirlo es configurar la profundidad del árbol o realizar poda, sin embargo esta última no está implementada en sklearn.

# E) SVM no lineal

A continuación se crea una svm no lineal para realizar la regresión. En este caso los hiper-parámetros utilizados son los por defecto, que corresponden  a $C=1$ y $epsilon=0.1$. Se utilizan además dos técnicas de pre procesamiento antes de utilizar el modelo, que corresponden a utilizar variables "dummy", es decir vectores binarios para las variables categóricas, y además se estandarizan los datos, ya que como se sabe, SVM y otros algoritmos de aprendizaje dependen de la escala que presentan los datos, por el uso de la distancia, entonces algunas variables pueden influir más por la naturaleza de sus valores, con lo cual el modelo puede presentar un mal desempeño en la regresión.

```python
df['cday'] = pd.to_datetime(df['datetime']).dt.dayofweek#0:lunes,6:domingo
df['cday'] = pd.to_numeric(df['cday'])
df['year'] = pd.to_datetime(df['datetime']).dt.year
df['year'] = pd.to_numeric(df['year'])

Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','year']]
Xdf = pd.get_dummies(Xdf,columns=['season', 'weather','hour','cday','year'])
Ydf=np.log(df.ix[:,'count'])

X_train = Xdf[0:ntrain].values
X_val = Xdf[ntrain:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:ntrain].values
Y_val = Ydf[ntrain:nval].values
Y_test = Ydf[nval:].values

from sklearn.preprocessing import StandardScaler
scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_val = scalerX.fit_transform(X_val)
X_test = scalerX.transform(X_test)

from sklearn.svm import SVR
model = SVR()
model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_val = model.predict(X_val)
Y_pred_test = model.predict(X_test)

kagg_train = eval_bikemodel(np.exp(Y_pred_train),np.exp(Y_train))
kagg_val = eval_bikemodel(np.exp(Y_pred_val),np.exp(Y_val))
kagg_test = eval_bikemodel(np.exp(Y_pred_test),np.exp(Y_test))

print "Modelo SVR utilizando mejoras en la representacion"
print "KAGG EVAL TRAIN =%f"%kagg_train
print "KAGG EVAL VAL =%f"%kagg_val
print "KAGG EVAL TEST =%f"%kagg_test
print "---------------------------------------------------------------------"
```

Inmediatamente se nota un gran incremento de calidad en la solución, ya que el árbol de regresión con la mejora en la representación reporta un KAGG error de 0.449708, mientras que SVR, disminuye este error a un valor de 0.338557, lo cual es una mejora muy significativa. El principal factor que se debe tener en cuenta, es que el problema es altamente no lineal, por lo mismo, los arboles de clasificación solo dividen el espacio en busca de reglas, lo cual hace que su desempeño sea limitado en este problema, mientras que SVR al mapear el espacio de características a un espacio de dimensión mucho más grande a través de una función no lineal, denominada kernel, que en este caso corresponde a radial basis function(rbf), puede realizar una regresión de los datos mucho más fácilmente mediante un híper-plano en el espacio de alta dimensionalidad, con lo cual es perfecto para este problema en particular.

Se deriva además que SVR generaliza de buena forma y exhibe los mejores resultados, incluso con los parámetros por defecto.

# F) Mejora de la SVM ajustando híper-parámetros

Se prueban diferentes valores de los parámetros $C$ y $epsilon$ con el fin de seguir disminuyendo el error. Se utilizan parámetros en diferentes escalas y diferentes tipos de kernel, con el fin de buscar los valores adecuados.

```python
C = [0.01, 0.1, 0.9, 1.2, 1.5, 1.8, 2.1]
E = [0.05,0.1]
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
kagg_list_val = []
parameter_list =[]
print "Kagg Error:"
for c in C:
    print "-"*80
    for e in E:
        model = SVR(C=c, epsilon = e)
        model.fit(X_train,Y_train)
        Y_pred_val = model.predict(X_val)
        kagg_val = eval_bikemodel(np.exp(Y_pred_val),np.exp(Y_val))
        print "KAGG VAL: %f"%kagg_val, ", C: ", c , ", epsilon: ", e
        kagg_list_val.append(kagg_val)
        parameter_list.append((c,e))

print "-"*80 + "\n"

minInd = kagg_list_val.index(min(kagg_list_val))

# Ahora se comparan kernel para los mejores valores  
print "Analisis de Kernel \n"
for k in kernel:
    print "Kernel: ", k
    model = SVR(C=parameter_list[minInd][0], epsilon = parameter_list[minInd][1], kernel = k)
    model.fit(X_train,Y_train)
    Y_pred_val = model.predict(X_val)
    kagg_val = eval_bikemodel(np.exp(Y_pred_val),np.exp(Y_val))
    print "KAGG VAL: %f"%kagg_val, ", C: ", parameter_list[minInd][0] , ", epsilon: ", parameter_list[minInd][1]
    print "-"*80

```

El parámetro $C$ , es el término que determina que tanto error se permite en la regresión. Para valores grandes de C, el modelo elige un margen pequeño, permitiendo así que más datos de entrenamiento ajusten correctamente. Por otra parte, para valores pequeños, la regularización es mayor, provocando que el margen sea grande, con lo cual más datos quedaran lejos de la línea o plano de regresión.

Con respecto a épsilon, este determina el ancho de la banda desde el híper-plano, en la cual se ignoran errores, entonces el modelo depende solo de los datos que estén fuera de esta banda.

Los mejores parámetros encontrados en este caso son $C =1.8$ , $epsilon =0.1$ con un kernel rbf. El valor de KAGG error  en el set de validación encontrado es de 0.304127.

Además, el valor KAGG en el set de pruebas para los mejores parámetros encontrados en el set de validación es:

```python
model = SVR(C=1.8 , epsilon=0.1, kernel="rbf")
model.fit(X_train,Y_train)
Y_pred_test = model.predict(X_test)
kagg_test = eval_bikemodel(np.exp(Y_pred_test),np.exp(Y_test))
print "KAGG EVAL TEST =%f"%kagg_test
```

Este valor, comparado al anterior sin configuración de parámetros, el cual era de 0.338557 no presenta mucha mejora, esto es quizás debido a que los parámetros por defecto funcionan muy bien. Una acotación es que al modificar $epsilon$, se presentaban peores resultados, sin embargo la configuración de $C$ es la que presento mejoras. Con respecto al kernel, rbf es el más adecuado en este caso, aunque otras opciones no lineales como kernel polinomial igualmente exhibe resultados favorables.

# G) Uso de Cross-Validation

Se utiliza Cross-Validation para encontrar los mejores parámetros, pero en este caso se utiliza el set de validación como parte del set de entrenamiento. Esto presumiblemente debe mejorar los resultados, ya que se presenta una mayor cantidad de datos para entrenar los modelos.

Primero se realiza cross-validation para el árbol de regresión. Se debe notar que se utiliza un rango de valores de max_depth = 10...20 y un rango de min_leaf = 1...10.

```python
from sklearn import cross_validation

Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','year']]
Ydf=np.log(df.ix[:,'count'])

X_train = Xdf[0:nval].values
#X_val = Xdf[ntrain:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:nval].values
#Y_val = Ydf[ntrain:nval].values
Y_test = Ydf[nval:].values

max_depth = range(10,20)
min_leaf = range(1,10)

parameter_list =[]
print "K-Fold Cross Validation Decision Tree \n\n"
k_fold = cross_validation.KFold(len(X_train),10)
best_cv =float("inf")
for md in max_depth:
        for ml in min_leaf:
            score_cv = 0
            for k, (train, val) in enumerate(k_fold):
                model = Tree(random_state=0,min_samples_leaf=ml, max_depth=md)
                model.fit(X_train[train], Y_train[train])
                Ypred_val = model.predict(X_train[val])
                Ytrue_val = Y_train[val]
                score_fold = eval_bikemodel(np.exp(Ypred_val),np.exp(Ytrue_val))
                score_cv += score_fold
            score_cv = score_cv / 10
            if score_cv < best_cv:
                best_cv = score_cv
                parameter_list.append((md,ml))
                print "Mejor KAGG: %f  Parametros: Max_Depth: %i , Min_Samples_Leaf: %i"%(score_cv, md, ml)

print "\nEND\n"
print "-"*80

# Test value
model = Tree(random_state=0,min_samples_leaf=parameter_list[-1][1], max_depth=parameter_list[-1][0])
model.fit(X_train,Y_train)
Y_pred_test = model.predict(X_test)
kagg_test = eval_bikemodel(np.exp(Y_pred_test),np.exp(Y_test))
print "KAGG EVAL TEST =%f"%kagg_test
```

En el resultado anterior se muestran los KAGG error, cada vez que se presenta una mejora utilizando cross validation, y sus respectivos parámetros. Finalmente se muestra el KAGG error del set de pruebas. 

En este caso, el error sobre el set de pruebas es 0.426343, anteriormente utilizando la misma representación pero sin cross validation  se obtenía un KAGG error de 0.449708. Esta mejora es de aproximadamente un $2.37\%$, lo cual es bastante considerando el hecho que solamente se utiliza el set de validación como entrenamiento y se usa cross validation, lo cual no aumenta la complejidad del modelo, solo son heurísticas para mejorar el desempeño. Los parámetros para el mejor modelo encontrado son Max_Depth: 14 y Min_Samples_Leaf: 3.

A continuación se realiza el mismo ejercicio pero ahora con SVR. El parámetro a evaluar en este caso es solamente $C$, ya que la influencia de $epsilon$ no es demasiada en este caso particular. Además se opta por acotar $C$ entre $0.75$ y $2$, ya que valores menores incrementan mucho el error de la regresión, y valores más grandes sobre ajustan mucho. Este comportamiento se percibió en el ítem anterior, es por ello que se tomó esta decisión.

```python
Xdf = pd.get_dummies(Xdf,columns=['season', 'weather','hour','cday','year'])
X_train = Xdf[0:nval].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:nval].values
Y_test = Ydf[nval:].values

scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_test = scalerX.transform(X_test)

# En este caso solo se ajusta parametro C, ya que epsilon presenta mejores resultados con 0.1
parameter_list =[]
print "K-Fold Cross Validation SVR \n\n"
Cs = np.linspace(0.75, 2, num=10)
k_fold = cross_validation.KFold(len(X_train),10)
best_cv =float("inf")
for c in Cs:
    score_cv = 0
    for k, (train, val) in enumerate(k_fold):
        model = SVR(C=c)
        model.fit(X_train[train], Y_train[train])
        Ypred_val = model.predict(X_train[val])
        Ytrue_val = Y_train[val]
        score_fold = eval_bikemodel(np.exp(Ypred_val),np.exp(Ytrue_val))
        score_cv += score_fold
    score_cv = score_cv / 10
    if score_cv < best_cv:
        best_cv = score_cv
        parameter_list.append(c)
        print "Mejor KAGG: %f  Parametros: C: %.3f , epsilon: 0.1"%(score_cv, c)

print "\nEND\n"
print "-"*80

# Test value
model = SVR(C=parameter_list[-1])
model.fit(X_train,Y_train)
Y_pred_test = model.predict(X_test)
kagg_test = eval_bikemodel(np.exp(Y_pred_test),np.exp(Y_test))
print "KAGG EVAL TEST =%f"%kagg_test
```

Nuevamente se presenta una mejora, que con respecto a los anteriores parámetros, que reportaban un error de 0.338545, disminuye en $1.47\%$. La conclusión que se obtiene entonces, es que es mejor utilizar cross validation sobre los datos (entrenamiento + validación) en vez de solo utilizar el set se entrenamiento y luego validar en el set de validación, ya que así se tiene más datos para entrenar y al vez se reduce el riesgo de overfitting.

El mejor parámetro es $C = 1.583$, que es efectivamente dentro del rango esperado, ya que en el ítem anterior, $C$ toma un valor de 1.8.

# H) Ensamblado de dos máquinas.

Para el desarrollo del ensamblado, se desarrollan dos máquinas, una que predice el atributo $\textit{casual}$, es decir el número de rentas iniciadas de usuarios no registrados en cierta hora, mientras que la otra, $\textit{registered}$, predice la de los usuarios registrados. Al sumar ambas variables, se obtiene la variable objetivo general del estudio, que corresponde a $\textit{count}$, por lo mismo para realizar el ensamblado de la regresión se ajustan los modelos, luego, se predice cada uno en el set de pruebas y se suman de manera lineal. Esto equivale a la predicción del modelo completo.

Para realizar el ensamblado, se utilizan dos SVR, ya que son las que demuestran los mejores resultados, además se utilizan los parámetros anteriormente encontrados. Se debe considerar además, que la transformación logaritmo de las variables objetivo debe ser corregida, ya que en este caso el valor real puede ser cero, provocando así que el logaritmo genere un valor de "-inf". Para corregirla entonces, se suma uno y luego antes de evaluar en la función KAGG, se resta uno.

```python
Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','year']]
Xdf = pd.get_dummies(Xdf,columns=['season', 'weather','hour','cday','year'])

# En este caso se suma 1 ya que casual y registered pueden tomar valor 0, log(0)= -inf
Ydf=np.log(df.ix[:,'count'])
Ydf1=np.log(df.ix[:,'casual']+1)
Ydf2=np.log(df.ix[:,'registered']+1)

Y_train = Ydf[0:nval].values
Y_test = Ydf[nval:].values

X_train = Xdf[0:nval].values
X_test = Xdf[nval:].values
Y_train1 = Ydf1[0:nval].values
Y_test1 = Ydf1[nval:].values

Y_train2 = Ydf2[0:nval].values
Y_test2 = Ydf2[nval:].values

scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_test = scalerX.transform(X_test)

model = SVR(C=1.583 , epsilon=0.1)
model.fit(X_train, Y_train1)
Y_pred_test1 = model.predict(X_test)

model = SVR(C=1.583 , epsilon=0.1)
model.fit(X_train, Y_train2)
Y_pred_test2 = model.predict(X_test)

Y_pred_test = (np.exp(Y_pred_test1)-1 )+ (np.exp(Y_pred_test2)-1)

kagg_test = eval_bikemodel(Y_pred_test,np.exp(Y_test))
print "KAGG EVAL TEST =%f"%kagg_test
```

El ensamblado de las SVR logra mejorar el KAGG anterior que correspondía a 0.323806. Una de las razones principales, es que se logra agregar diversidad al manipular los datos de entrada, ya que cada máquina puede especializarse en un problema, en este caso usuarios registrados y no registrados que arriendan bicicletas. Entonces al sumar ambas predicciones, se logra un mejor resultado que el modelo completo, que resulta ser más complejo.

Así, usar ensamblados es útil en este caso, ya que los datos lo permiten, debido a que se agrega diversidad y se dividen las tareas. Otra forma de agregar diversidad serie re muestreando los datos o generando sub muestras por ejemplo que consideren solo ciertos atributos, como día y año, o como temperatura y humedad. Esto puede agregar aún más diversidad y en el mejor de los casos mejorar el desempeño del ensamblado.

# i) Evaluación ensamblados: número de máquinas vs rendimiento

Para evaluar que tanto afecta el número de máquinas al rendimiento, se estudia un tipo de ensamblado muy común, que corresponde a Random Forest para regresión en este caso. Este algoritmo es una adaptación de Bagging que utiliza un número determinado de árboles para realizar las predicciones.

Random Forest en regresión simplemente promedia los resultados obtenidos de cada árbol, y la idea de este es mejorar la generalización y aumentar la robustez del modelo.

Para el estudio, se generan de 2 a 100 árboles, utilizando los mejores parámetros anteriormente encontrados (max_depth=14, min_samples_leaf=3), y se grafica el error de entrenamiento y de pruebas. Con esto se puede estudiar el efecto que tiene el número de máquinas en el desarrollo del ensamblado.

```python
Xdf=df.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday','year']]
Ydf=np.log(df.ix[:,'count'])

X_train = Xdf[0:ntrain].values
X_test = Xdf[nval:].values
Y_train = Ydf[0:ntrain].values
Y_test = Ydf[nval:].values

from sklearn.ensemble import RandomForestRegressor

machines = range(2, 100)
results_train = []
results_test = []
for n in machines:
    model = RandomForestRegressor(n_estimators=n,max_depth=14, min_samples_leaf=3,random_state=0)
    model.fit(X_train, Y_train)

    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    kagg_train = eval_bikemodel(np.exp(Y_pred_train),np.exp(Y_train))
    kagg_test = eval_bikemodel(np.exp(Y_pred_test),np.exp(Y_test))
    results_train.append(kagg_train)
    results_test.append(kagg_test)

imin_train = results_train.index(min(results_train))
imin_test = results_test.index(min(results_test))

plt.figure(figsize=(20,10))
plt.plot(machines, results_train, "r", label="KAGG train")
plt.plot(machines, results_test, "b", label="KAGG test")
plt.xlabel("Numero de maquinas")
plt.ylabel("Kagg Error")
plt.title("Numero de maquinas vs Error en Random Forest")
plt.legend()
plt.show()

print "MIN KAGG TRAIN: %f , NUM MAQUINAS: %i"%(results_train[imin_train], imin_train+ 2)
print "MIN KAGG TEST: %f , NUM MAQUINAS: %i"%(results_test[imin_test], imin_test+ 2)
```

Al observar el grafico anterior, es claro el comportamiento que presenta Random Forest. En un principio, cuando el número de máquinas es pequeño, el error es alto, luego al agregar maquinas el error decae significativamente, hasta llegar a un punto en donde el decaimiento se torna lento y paulatino. Esto puede ser explicado por el hecho de que Random Forest al usar sub muestras de los datos de entrenamiento y además al utilizar un subconjunto aleatorio de entre todas las características para hacer las divisiones de los nodos, provoca que se incremente el sesgo, pero al promediar los resultados de cada árbol, la varianza disminuye en mayor medida, lo que provoca esta alta disminución del error. 

Luego cuando se agregan más máquinas, la ganancia se torna más difícil, ya que la diversidad que aporta cada nueva máquina es baja, provocando el comportamiento que se observa en el gráfico, en donde a partir de cierto punto la ganancia en términos de precisión es muy baja comparada con las máquinas y el tiempo de computación que conlleva crear el ensamblado produciendo así un trade-off de ganancia/tiempo computacional.

Con respecto a la generalización, los errores tanto de entrenamiento y pruebas se comportan similar, es decir, siguen un mismo patrón, pero con una diferencia en el error, ya que el ensamblado en este caso no produce generalización en mayor grado, solamente reduce el error simultáneamente en ambos sets.

Un detalle a tener en cuenta es que el error siempre sigue disminuyendo, pero el beneficio es tan poco, que se torna infactible adicionar más máquinas. Para determinar cuántas máquinas son apropiadas se puede establecer una tolerancia al error. En este caso a partir de aproximadamente 30 máquinas la mejora no es significativa.
